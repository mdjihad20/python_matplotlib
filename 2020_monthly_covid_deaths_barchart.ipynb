```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
year2020_months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"] 
death_cases = [0,5,4326,55315,41137,19475,25249,30244,23329,23545,37065,77112] 
#January month is 0 or N/A bcz covid started from FEB
```

```python
yaxis = np.arange(len(year2020_months)) #We need list numbers to plot , string can't be plotted!
yaxis
```

```python
plt.xticks(yaxis,year2020_months) #It replaces (x-axis) indexes to it's value (name) like: for 0 index it's value is "Jan"

plt.xlabel("Month's")
plt.ylabel("Death Tolls")
plt.title("Covid 19 Death Cases of 2020 ")
plt.bar(yaxis, death_cases, label="Deaths" , color="#1B93D7")

plt.legend()
```

```python
plt.yticks(yaxis,year2020_months) #It replaces (y-axis) indexes to it's value (name) like: for 0 index it's value is "Jan"

plt.xlabel("Death Tolls")
plt.ylabel("Month's")
plt.title("Covid 19 Death Cases of 2020 ")
plt.barh(yaxis, death_cases, label="Deaths" , color="#1B93D7") # plt.barh for horizontal bar-chart

plt.legend()
```
