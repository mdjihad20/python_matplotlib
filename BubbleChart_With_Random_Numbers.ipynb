```python
#  Important library and packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(70)
y = np.random.rand(70)
z = np.random.rand(70)
colors = np.random.rand(70)
```

```python
# Using scatter plot function 
plt.scatter(x, y, s = z*1000, c = colors)
plt.show()
```

```python
# With less rand values
x = np.random.rand(30)
y = np.random.rand(30)
z = np.random.rand(30)
colors = np.random.rand(30)

# Using scatter plot function 
plt.scatter(x, y, s = z*1000, c = colors)
plt.show()
```
