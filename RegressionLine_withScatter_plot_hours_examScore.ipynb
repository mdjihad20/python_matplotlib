```python
import matplotlib.pyplot as plt
import seaborn 
import pandas as pd
```

```python
#data   #x , y
study_hours = pd.Series(data=[5, 6, 6, 7, 10, 11, 9, 4, 2, 6, 10, 3, 8, 6.5, 6])
exam_score  = pd.Series(data=[78, 79, 77, 80, 99, 88, 93, 70, 53, 80, 98, 71, 88, 83, 82])
```

```python
# plot scatter 
plt.scatter(x = study_hours, y = exam_score)

# titile and grid

plt.title("Study Hours / Exam Score Chart")
plt.xlabel("Hours spent studying")
plt.ylabel("Exam Score out of 100")
```

```python
# Scatter Plot With Regression Line
seaborn.regplot(x = study_hours, y = exam_score)
```
