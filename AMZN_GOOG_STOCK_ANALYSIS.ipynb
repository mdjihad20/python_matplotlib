```python
# Import packages and modules !
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
# Get data from pandas_datareader , stocks name = "GOOG" , Starting date and Closing price that day!
data = pdr.get_data_yahoo(['GOOG', 'AMZN'], start= '2022-01-01')['Close']
data.head()
```

```python
# Scatter plot closing prices!
plt.scatter(data.GOOG, data.AMZN , alpha = 0.7)
```

```python
returns = np.log(data).diff()
returns.head()
returns.dropna()
```

```python
plt.axvline(0, c=(.5, .5, .5, .5), ls='--')
plt.axhline(0, c=(.5, .5, .5, .5), ls='--')

plt.style.use('seaborn')
plt.scatter(returns.GOOG, returns.AMZN, c = returns.GOOG, cmap = 'Spectral', edgecolor='k')
plt.colorbar()
```
