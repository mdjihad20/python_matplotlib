```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
military_spending = [801, 293, 77, 68, 66]
countries = ["USA", "China", "India", "United Kingdom", "Russia"]
```

```python
plt.title("World Top 5 Countries For Military Spendings (In Billions)")
plt.axis("equal")
plt.pie(military_spending, 
        labels = countries,
        radius = 1.1,
        autopct = '%0.f%%',# For decimel 0.2f for 61.34 percent like this
        explode = [0,0,0.05,0.05,0.05],
        startangle = 45) 
plt.legend()
plt.show()
```
