```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mens_weight_10per   = [49 , 52 , 50 , 55 , 68 , 70 , 76 , 72 , 54 , 59]
womens_weight_10per = [45 , 48 , 50 , 53 , 65 , 70 , 69 , 71 , 51 , 53]
```

```python
plt.xlabel("Weight (KG)")
plt.ylabel("Persons")
plt.title("10 Men & Women Weights (Age between 18-20) - Histogram Chart")

plt.hist(
    [mens_weight_10per,womens_weight_10per], 
    bins=[45,50,55,60,65,70,75,80], 
    rwidth=0.95 , color = ['#103561','#AF0272'],
    label = ["Men","Women"]) 
plt.legend()


#without second parametre plt.hist() creates 10 bins and arrange itself 
```

```python
plt.hist(
    [mens_weight_10per,womens_weight_10per], 
    bins=[45,50,55,60,65,70,75,80], 
    rwidth=0.95 , color = ['#103561','#AF0272'],
    label = ["Men","Women"],
    orientation = "horizontal") # (orientation = "horizontal)"
plt.legend()
```
